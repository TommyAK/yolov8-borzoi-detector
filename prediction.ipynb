```python
from ultralytics import YOLO
import cv2
import time
```

```python
# selecting best trained model
model = YOLO("C:\\Users\\Tommy\\Documents\\yolo\\borzoi classifier\\runs\\detect\\train\\weights\\best.pt")
```

```python
# predict test images
model.predict(conf=0.25, source="C:\\Users\\Tommy\\Documents\\yolo\\borzoi classifier\\data\\test\\images", save=True)
```

```python
# predict test videos (sourced from YouTube)
model.predict(conf=0.25, source="C:\\Users\\Tommy\\Documents\\yolo\\borzoi classifier\\data\\unseen_videos", save=True)
```

```python
# set video path
video_path = "C:\\Users\\Tommy\\Documents\\yolo\\borzoi classifier\\data\\unseen_videos\\Young Borzoi asking old cat to play [_aK39LG6Y1s].mp4"

# intialise opencv video pathcapture
vid = cv2.VideoCapture(video_path)

# open video as process each frame
while vid.isOpened():
    success, frame = vid.read()

    if success:
        # saving results for specific frame
        results = model(frame)

        # annotate frame with bounding box
        annotated_frame = results[0].plot()

        # open external window to preview result
        cv2.imshow("Preview Window", annotated_frame)
        
        # alter rate each frame appears

        time.sleep(0.005)
        # stop early key
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

vid.release()
cv2.destroyAllWindows()
```
